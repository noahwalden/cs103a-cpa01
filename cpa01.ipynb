# Video Game Sales Analysis

This dataset is called Video Game Sales. I downloaded it from https://www.kaggle.com/gregorut/videogamesales. It is a dataset containing the sales numbers, by region, of ~16k games that sold more than 100,000 copies each. You can download the dataset by going to the link and clicking the "Download" button in the top right of the page.

Questions:
- Which platform has the greatest success in terms of global sales?
- Which year saw the greatest number of sales by region?
- When the game "Call of Duty: Black Ops" released, how did it perform on different consoles in each of the regions whose data is available in the dataset?
- How did certain major publishers' global sales compare to one another over the course of the time period spanned by the data set?


## Data Visualizations

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
sales_df = pd.read_csv("data/vgsales.csv")
sales_df
```

```python
sales_df.describe()
```

The index list just contains the indices of the games in the data sets, which means that it isn't very interesting or informative.

```python
list(sales_df.index)
```

```python
list(sales_df.columns)
```

```python
title = "'Call of Duty: Black Ops' Global Sales (in millions) by Console"
sales_df[sales_df["Name"] == "Call of Duty: Black Ops"].plot(kind="bar", x="Platform", y=["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"], title=title, figsize=(10,5))
```

```python
df4 = sales_df.groupby(['Year'])[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]].agg(sum)
title = "Regional Sales (in millions) by Year"
df4.plot(title=title, figsize=(10,5))
```

```python
title = "Total Global Sales (in millions) For All Consoles"
df5 = sales_df.groupby(['Platform']).Global_Sales.agg(sum)
df5.plot(kind="bar", figsize=(10,5))
```

```python
pt = pd.pivot_table(sales_df, aggfunc=sum, index="Year", columns="Publisher", values="Global_Sales")[["Activision", "Nintendo", "Sony Computer Entertainment", "Sega"]]
pt
```

```python
title = "Total Global Sales by Year for Four Major Publishers"
pt.plot(figsize=(10,5), title=title)
```

## Analysis

At the beginning of this assignment, I asked the following questions:

1. Which platform has the greatest success in terms of global sales?
2. Which year saw the greatest number of sales by region?
3. When the game "Call of Duty: Black Ops" released, how did it perform on different consoles in each of the regions whose data is available in the dataset?
4. How did certain major publishers' global sales compare to one another over the course of the time period spanned by the data set?

The platform that had the greatest success in terms of global sales was by far the PS2, followed by the Xbox 360. Even though the PC platform is becoming much more popular, I think that the PS2 does so well in this dataset because it appeared at a time when there was a great deal of interest in video games AND indie game development and distribution was basically impossible. This means that the games that were produced had the backing of big publishers and were likely to succeed.

Right around 2006 is when we see the greatest number of sales by region, which is around the time when the Wii, a massively popular console, came out. I think that my analysis of this spike in sales can be attributed to Wii Sports, which caused the total global sales of Nintendo to spike around the same time. 

When Call of Duty: Black Ops released, it sold the most copies in North America on the Xbox 360. I think the most interesting thing to note about this plot is that the game didn't sell very well in Japan, but almost all of its copies in that region were sold for the PS3. This could prompt further questions about the popularities of certain consoles by region, which would require another dataset about console sales.
